```python
# Classify movies based on Rating
#  (Excellent, Good, Average):
data.columns
```

```python
data["Rating"]
```

```python
def movies_rater(rating):
    if rating >= 7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data["Rating_Category"]= data["Rating"].apply(movies_rater)
```

```python
list1= []
for value in data["Genre"]:
    list1.append(value.split(","))
```

```python
list1
```

```python
one_d= []
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list= []
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
